```python
data = pd.read_csv("viol_numero_denuncias_ok.csv") 
data.tail()

```

```python
##OLS numero de denuncias en funcion de la alcaldia
#Regresión solo por alcaldías: 
fit = ols("numero_denuncias ~ C(alcaldia)", data=data).fit()
fit.summary()
```

```python
denuncias=pd.read_csv("Denuncias_por_alcaldia.csv")
```

```python
denuncias.head(10)
```

```python
denuncias_2016_2019=denuncias.loc[(denuncias.anio<2020)]
```

```python
denuncias_2016_2019.tail(10)
```

```python
denuncias_2016_2019["poblacion"]=poblacion
```

```python
denuncias_2016_2019["denuncias_100mil_habitantes"]=denuncias_2016_2019.numero_denuncias/(denuncias_2016_2019.poblacion/100000)
```

```python
denuncias_2016_2019["pobreza"]=pobreza
```

```python
denuncias_2016_2019.to_csv("denuncias_2016_2019.csv")
```

```python
#hago una regresión con las denuncias por cada 100 mil habitantes de 2019 
#en relación a la alcaldía, el nivel de pobreza. 
```

```python
fit = ols("denuncias_100mil_habitantes ~ C(alcaldia)+pobreza", data=denuncias_2016_2019).fit()
fit.summary()
```
