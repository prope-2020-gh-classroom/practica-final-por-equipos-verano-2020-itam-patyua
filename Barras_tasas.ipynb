```python
data  = pd.read_csv("denuncias_2016_2019_cada_100mil.csv")
```

### Gráficas

```python
plot2019= data[data.anio == 2019]["denuncias_100mil_habitantes"].plot(kind="bar")
plt.xticks(np.arange(16), ("ALVARO OBREGON","AZCAPOTZALCO","BENITO JUAREZ","COYOACAN", "CUAJIMALPA DE MORELOS","CUAUHTEMOC","GUSTAVO A MADERO" ,"IZTACALCO", "IZTAPALAPA","LA MAGDALENA CONTRERAS","MIGUEL HIDALGO","MILPA ALTA","TLAHUAC","TLALPAN","VENUSTIANO CARRANZA","XOCHIMILCO"))
plt.ylabel("Carpetas de investigación ")
plt.xlabel("Alcaldía")
plt.title("Número carpetas de investigación por cada 100 mil habitantes 2019")
```

```python
plot2019= data[data.anio == 2016]["denuncias_100mil_habitantes"].plot(kind="bar")
plt.xticks(np.arange(16), ("ALVARO OBREGON","AZCAPOTZALCO","BENITO JUAREZ","COYOACAN", "CUAJIMALPA DE MORELOS","CUAUHTEMOC","GUSTAVO A MADERO", "IZTACALCO", "IZTAPALAPA","LA MAGDALENA CONTRERAS","MIGUEL HIDALGO","MILPA ALTA","TLAHUAC","TLALPAN","VENUSTIANO CARRANZA","XOCHIMILCO"))
plt.ylabel("Carpetas de investigación ")
plt.xlabel("Alcaldía")
plt.title("Número carpetas de investigación por cada 100 mil habitantes 2016")
```

Considerando los habitantes de cada alcaldía podemos ver que las diferencias se aplanan y las tasas más altas de denuncias corresponden a Iztacalco, Cuauhtémoc y la Magadalena Contreras.
